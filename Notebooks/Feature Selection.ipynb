Create new column positive words which gives number of positive words in reviews

```python
# What punctuation is included in the default list?
import string

string.punctuation
```

```python
# Define a function to remove punctuation in our messages
def remove_punct(text):
    text = "".join([char for char in text if char not in string.punctuation])
    return text
```

```python
df['reviews_list'] = df['reviews_list'].apply(lambda x: remove_punct(x))
```

```python
#import regular expression
import re

def tokenize(text):
    tokens = text.split()
    return tokens

df['reviews_list'] = df['reviews_list'].apply(lambda x: tokenize(x.lower()))

df['reviews_list']
```

```python
# Import the NLTK package and download the necessary data
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
ENGstopwords = stopwords.words('english')
```

```python
ENGstopwords = ENGstopwords + ['rated','40','ratedn','10','0','20','30','50']
```

```python
# Define a function to remove all stopwords
def remove_stopwords(tokenized_text):    
    text = [word for word in tokenized_text if word not in ENGstopwords]
    return text

df['reviews_list'] = df['reviews_list'].apply(lambda x: remove_stopwords(x))
```

```python
df['reviews_list']
```

```python
#Lemmatization
# import these modules
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
  
lemmatizer = WordNetLemmatizer() 
```

```python
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = []
    for word in text:
        lemma = lemmatizer.lemmatize(word)
        lemmatized_words.append(lemma)
    return ' '.join(lemmatized_words)
```

```python
df['Review_clean'] = df['reviews_list'].apply(lambda x: lemmatize_text(x))
```

```python
import gensim

# Load Google's pre-trained Word2Vec model.
model = gensim.models.KeyedVectors.load_word2vec_format("C:/Users/patel/GoogleNews-vectors-negative300.bin.gz", binary=True)
```

```python
positive_seeds = ['good', 'great', 'excellent', 'wonderful', 'awesome']
positive_words = []

for seed_word in positive_seeds:
    similar_words = model.most_similar(positive=[seed_word], topn=500)
    positive_words.extend([word for word, similarity in similar_words])
```

```python
positive_words.remove('bad')
len(positive_words)
```

```python
positive_words = list(set(positive_words))
len(positive_words)
```

```python
positive_words
```

```python
from nltk.corpus import wordnet

def count_positive_words(review):
    count = 0
    tokens = tokenize(review)
    for token in tokens:
        synsets = wordnet.synsets(token)
        for synset in synsets:
            if synset.pos() == 'a' and synset.lemmas()[0].name() in positive_words:
                count += 1
    return count
```

```python
df['positive_words'] = df['Review_clean'].apply(count_positive_words)
```
